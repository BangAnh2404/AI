```python
#bai1
import heapq
import copy
TRANG_THAI_DICH = [
    [1, 2, 3],
    [8, 0, 4],
    [7, 6, 5]
]

class Nut:
    def __init__(self, bang, g, cha=None):
        self.bang = bang
        self.g = g  # Chi phí từ điểm bắt đầu đến hiện tại
        self.h = self.tinh_h()  # Heuristic: số ô sai vị trí
        self.f = self.g + self.h # Giá trị f(n)
        self.cha = cha

    def tinh_h(self):
        """Đếm số ô không nằm đúng vị trí (không tính ô trống số 0)"""
        dem = 0
        for i in range(3):
            for j in range(3):
                if self.bang[i][j] != 0 and self.bang[i][j] != TRANG_THAI_DICH[i][j]:
                    dem += 1
        return dem
    def __lt__(self, khac):
        return self.f < khac.f

    def lay_chuoi_bang(self):
        """Chuyển bảng thành tuple để lưu vào mảng CLOSE (dễ so sánh)"""
        return tuple(tuple(dong) for dong in self.bang)

def tim_vi_tri_trong(bang):
    for i in range(3):
        for j in range(3):
            if bang[i][j] == 0:
                return i, j

def lay_cac_nut_ke(nut_hien_tai):
    cac_nut_ke = []
    r, c = tim_vi_tri_trong(nut_hien_tai.bang)
    huong = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    for dr, dc in huong:
        r_moi, c_moi = r + dr, c + dc

        if 0 <= r_moi < 3 and 0 <= c_moi < 3:
            bang_moi = copy.deepcopy(nut_hien_tai.bang)
            bang_moi[r][c], bang_moi[r_moi][c_moi] = bang_moi[r_moi][c_moi], bang_moi[r][c]

            moi = Nut(bang_moi, nut_hien_tai.g + 1, nut_hien_tai)
            cac_nut_ke.append(moi)

    return cac_nut_ke
def nhap_bang():
    """Hàm cho phép người dùng nhập bảng 3x3 từ bàn phím"""
    print("--- Nhập trạng thái bảng 3x3 (nhập số 0 cho ô trống) ---")
    bang = []
    for i in range(3):
        while True:
            try:
                dong_str = input(f"Nhập hàng {i + 1} (3 số cách nhau bởi dấu cách, enter khi đủ 3 số): ")
                dong = [int(x) for x in dong_str.split()]

                if len(dong) != 3:
                    print("Lỗi: Bạn phải nhập đúng 3 số!")
                    continue

                bang.append(dong)
                break
            except ValueError:
                print("Lỗi: Vui lòng chỉ nhập số nguyên!")
    return bang

def giai_8_puzzle(bang_dau):

    nut_bat_dau = Nut(bang_dau, 0)
    open_list = []
    heapq.heappush(open_list, nut_bat_dau)


    close_list = set()

    print("Đang tìm lời giải...")

    while open_list:

        remove_node = heapq.heappop(open_list)


        if remove_node.h == 0:
            print("Tìm thấy lời giải!")
            in_ket_qua(remove_node)
            return


        trang_thai_hien_tai = remove_node.lay_chuoi_bang()
        close_list.add(trang_thai_hien_tai)

        for ke in lay_cac_nut_ke(remove_node):
            if ke.lay_chuoi_bang() not in close_list:
                heapq.heappush(open_list, ke)



    print("Không tìm thấy đường đi.")

def in_ket_qua(nut):
    duong_di = []
    hien_tai = nut
    while hien_tai:
        duong_di.append(hien_tai.bang)
        hien_tai = hien_tai.cha

    for i, bang in enumerate(reversed(duong_di)):
        print(f"Bước {i}:")
        for dong in bang:
            print(dong)
        print("---")


if __name__ == "__main__":

    bang_khoi_dau = nhap_bang()


    print("\nBảng bạn đã nhập:")
    for dong in bang_khoi_dau:
        print(dong)
    print("------------------")

    giai_8_puzzle(bang_khoi_dau)
```
