```python
import heapq

class Nut:
    def __init__(self, duong_di, thanh_pho_hien_tai, cac_thanh_pho_chua_tham, chi_phi_g, chi_phi_h, cha):
        self.duong_di = duong_di
        self.thanh_pho_hien_tai = thanh_pho_hien_tai
        self.cac_thanh_pho_chua_tham = cac_thanh_pho_chua_tham
        self.chi_phi_g = chi_phi_g
        self.chi_phi_h = chi_phi_h
        self.chi_phi_f = chi_phi_g + chi_phi_h
        self.cha = cha

    def __lt__(self, other):
        return self.chi_phi_f < other.chi_phi_f

def ham_heuristic(thanh_pho_hien_tai, cac_thanh_pho_chua_tham, khoang_cach, so_thanh_pho, thanh_pho_bat_dau):
    chi_phi_h = 0
    khong_co_gioi_han = float('inf')

    if not cac_thanh_pho_chua_tham:
        return khoang_cach[thanh_pho_hien_tai][thanh_pho_bat_dau]

    min_den_tiep_theo = khong_co_gioi_han
    for tp_chua_tham in cac_thanh_pho_chua_tham:
        min_den_tiep_theo = min(min_den_tiep_theo, khoang_cach[thanh_pho_hien_tai][tp_chua_tham])
    chi_phi_h += min_den_tiep_theo

    min_tro_ve_dau = khong_co_gioi_han
    for tp_chua_tham in cac_thanh_pho_chua_tham:
        min_tro_ve_dau = min(min_tro_ve_dau, khoang_cach[tp_chua_tham][thanh_pho_bat_dau])
    chi_phi_h += min_tro_ve_dau

    return chi_phi_h

def thuat_toan_astar(khoang_cach):
    so_thanh_pho = len(khoang_cach)
    thanh_pho_bat_dau = 0

    cac_thanh_pho_chua_tham_khoi_tao = frozenset(range(1, so_thanh_pho))

    nut_khoi_dau = Nut(
        duong_di=(thanh_pho_bat_dau,),
        thanh_pho_hien_tai=thanh_pho_bat_dau,
        cac_thanh_pho_chua_tham=cac_thanh_pho_chua_tham_khoi_tao,
        chi_phi_g=0,
        chi_phi_h=ham_heuristic(thanh_pho_bat_dau, cac_thanh_pho_chua_tham_khoi_tao, khoang_cach, so_thanh_pho, thanh_pho_bat_dau),
        cha=None
    )

    hang_doi_uu_tien = [nut_khoi_dau]
    tap_da_duyet = { (thanh_pho_bat_dau, cac_thanh_pho_chua_tham_khoi_tao) }

    while hang_doi_uu_tien:
        nut_hien_tai = heapq.heappop(hang_doi_uu_tien)

        if not nut_hien_tai.cac_thanh_pho_chua_tham:
            chi_phi_cuoi_cung = khoang_cach[nut_hien_tai.thanh_pho_hien_tai][thanh_pho_bat_dau]
            if chi_phi_cuoi_cung == float('inf'):
                return None, None
            tong_chi_phi = nut_hien_tai.chi_phi_g + chi_phi_cuoi_cung
            duong_di_cuoi_cung = nut_hien_tai.duong_di + (thanh_pho_bat_dau,)
            return duong_di_cuoi_cung, tong_chi_phi

        for thanh_pho_ke_tiep in list(nut_hien_tai.cac_thanh_pho_chua_tham):
            chi_phi_canh = khoang_cach[nut_hien_tai.thanh_pho_hien_tai][thanh_pho_ke_tiep]
            if chi_phi_canh == float('inf'):
                continue

            chi_phi_g_moi = nut_hien_tai.chi_phi_g + chi_phi_canh
            duong_di_moi = nut_hien_tai.duong_di + (thanh_pho_ke_tiep,)
            cac_thanh_pho_chua_tham_moi = frozenset(tp for tp in nut_hien_tai.cac_thanh_pho_chua_tham if tp != thanh_pho_ke_tiep)
            chi_phi_h_moi = ham_heuristic(thanh_pho_ke_tiep, cac_thanh_pho_chua_tham_moi, khoang_cach, so_thanh_pho, thanh_pho_bat_dau)

            trang_thai_moi = (thanh_pho_ke_tiep, cac_thanh_pho_chua_tham_moi)

            if trang_thai_moi not in tap_da_duyet:
                tap_da_duyet.add(trang_thai_moi)
                nut_moi = Nut(
                    duong_di=duong_di_moi,
                    thanh_pho_hien_tai=thanh_pho_ke_tiep,
                    cac_thanh_pho_chua_tham=cac_thanh_pho_chua_tham_moi,
                    chi_phi_g=chi_phi_g_moi,
                    chi_phi_h=chi_phi_h_moi,
                    cha=nut_hien_tai
                )
                heapq.heappush(hang_doi_uu_tien, nut_moi)

    return None, None

def in_duong_di(duong_di):
    if not duong_di:
        return
    chuoi_duong_di = ""
    for i, tp in enumerate(duong_di):
        chuoi_duong_di += str(tp)
        if i < len(duong_di) - 1:
            chuoi_duong_di += " -> "
    print(chuoi_duong_di)
khong_co_gioi_han = float('inf')

# Vi du ma tran khoang cach (4 thanh pho)
# thanh pho 0 -> A, 1 -> B, 2 -> C, 3 -> D
khoang_cach_4_thanh_pho = [
    [khong_co_gioi_han, 10, 15, 20],
    [10, khong_co_gioi_han, 35, 25],
    [15, 35, khong_co_gioi_han, 30],
    [20, 25, 30, khong_co_gioi_han]
]

print("Ma tran khoang cach:")
for hang in khoang_cach_4_thanh_pho:
    print(hang)
print()
duong_di_toi_uu, tong_chi_phi_toi_uu = thuat_toan_astar(khoang_cach_4_thanh_pho)

if duong_di_toi_uu:
    print("Duong di toi uu:")
    in_duong_di(duong_di_toi_uu)
    print(f"Tong chi phi: {tong_chi_phi_toi_uu}")
else:
    print("Khong tim thay duong di.")

```
