```python
import heapq

def thuat_toan_a_star(do_thi, toa_do, diem_bat_dau, diem_ket_thuc):
    tap_mo = []
    heapq.heappush(tap_mo, (0, diem_bat_dau))
    luu_vet = {}
    chi_phi_g = {nut: float('inf') for nut in do_thi}
    chi_phi_g[diem_bat_dau] = 0
    chi_phi_f = {nut: float('inf') for nut in do_thi}
    chi_phi_f[diem_bat_dau] = tinh_heuristic(diem_bat_dau, diem_ket_thuc, toa_do)
    while tap_mo:
        _, nut_hien_tai = heapq.heappop(tap_mo)
        if nut_hien_tai == diem_ket_thuc:
            return tai_thiet_lap_duong_di(luu_vet, nut_hien_tai)

        for lang_gieng, khoang_cach in do_thi[nut_hien_tai].items():

            g_tam_thoi = chi_phi_g[nut_hien_tai] + khoang_cach

            if g_tam_thoi < chi_phi_g[lang_gieng]:

                luu_vet[lang_gieng] = nut_hien_tai
                chi_phi_g[lang_gieng] = g_tam_thoi
                chi_phi_f[lang_gieng] = g_tam_thoi + tinh_heuristic(lang_gieng, diem_ket_thuc, toa_do)


                if lang_gieng not in [n[1] for n in tap_mo]:
                    heapq.heappush(tap_mo, (chi_phi_f[lang_gieng], lang_gieng))

    return None

def tinh_heuristic(nut_a, nut_b, toa_do):

    x1, y1 = toa_do[nut_a]
    x2, y2 = toa_do[nut_b]
    return ((x1 - x2)**2 + (y1 - y2)**2)**0.5

def tai_thiet_lap_duong_di(luu_vet, hien_tai):
    duong_di = [hien_tai]
    while hien_tai in luu_vet:
        hien_tai = luu_vet[hien_tai]
        duong_di.append(hien_tai)
    return duong_di[::-1]


do_thi_mau = {
    'A': {'B': 1, 'C': 3},
    'B': {'A': 1, 'D': 4, 'E': 2},
    'C': {'A': 3, 'F': 5},
    'D': {'B': 4},
    'E': {'B': 2, 'F': 1},
    'F': {'C': 5, 'E': 1}
}


toa_do_cac_nut = {
    'A': (0, 0), 'B': (1, 1), 'C': (2, 0),
    'D': (1, 5), 'E': (3, 2), 'F': (4, 0)
}

# Chay thu
ket_qua = thuat_toan_a_star(do_thi_mau, toa_do_cac_nut, 'A', 'F')
print(f"Duong di tim duoc: {' -> '.join(ket_qua)}")
```
