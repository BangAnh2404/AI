```python
may = 'X'
nguoi = 'O'

def in_bang(bang):
    print("-------------")
    for i in range(3):
        print(f"| {bang[i][0]} | {bang[i][1]} | {bang[i][2]} |")
        print("-------------")

def kiem_tra_thang(bang, nguoi_choi):
    # Kiem tra cac hang
    for i in range(3):
        if bang[i][0] == bang[i][1] == bang[i][2] == nguoi_choi:
            return True

    # Kiem tra cac cot
    for i in range(3):
        if bang[0][i] == bang[1][i] == bang[2][i] == nguoi_choi:
            return True

    # Kiem tra hai duong cheo
    if bang[0][0] == bang[1][1] == bang[2][2] == nguoi_choi:
        return True
    if bang[0][2] == bang[1][1] == bang[2][0] == nguoi_choi:
        return True

    return False

def het_cho_trong(bang):
    for i in range(3):
        for j in range(3):
            if bang[i][j] == ' ':
                return False
    return True

def tinh_diem(bang):
    if kiem_tra_thang(bang, may):
        return 10
    elif kiem_tra_thang(bang, nguoi):
        return -10
    else:
        return 0

def minimax(bang, do_sau, la_luot_cua_may):
    diem = tinh_diem(bang)

    # Neu may thang
    if diem == 10:
        return diem - do_sau

    # Neu nguoi thang
    if diem == -10:
        return diem + do_sau

    # Hoa
    if het_cho_trong(bang):
        return 0

    if la_luot_cua_may:
        diem_tot_nhat = -1000
        for i in range(3):
            for j in range(3):
                if bang[i][j] == ' ':
                    bang[i][j] = may
                    diem_hien_tai = minimax(bang, do_sau + 1, False)
                    bang[i][j] = ' '
                    diem_tot_nhat = max(diem_tot_nhat, diem_hien_tai)
        return diem_tot_nhat

    else:
        diem_tot_nhat = 1000
        for i in range(3):
            for j in range(3):
                if bang[i][j] == ' ':
                    bang[i][j] = nguoi
                    diem_hien_tai = minimax(bang, do_sau + 1, True)
                    bang[i][j] = ' '
                    diem_tot_nhat = min(diem_tot_nhat, diem_hien_tai)
        return diem_tot_nhat

def tim_nuoc_di_tot_nhat(bang):
    diem_tot_nhat = -1000
    nuoc_di = (-1, -1)

    for i in range(3):
        for j in range(3):
            if bang[i][j] == ' ':
                bang[i][j] = may
                diem_nuoc_di = minimax(bang, 0, False)
                bang[i][j] = ' '

                if diem_nuoc_di > diem_tot_nhat:
                    diem_tot_nhat = diem_nuoc_di
                    nuoc_di = (i, j)

    return nuoc_di

def luot_nguoi_choi(bang):
    while True:
        try:
            print("Nhap vi tri ban muon danh (tu 0 den 2).")
            hang = int(input("Nhap hang: "))
            cot = int(input("Nhap cot: "))

            # Kiem tra xem toa do co nam trong ban co khong
            if hang < 0 or hang > 2 or cot < 0 or cot > 2:
                print("Vi tri khong hop le! Vui long nhap so tu 0 den 2.\n")
                continue

            # Kiem tra xem o do da co nguoi danh chua
            if bang[hang][cot] == ' ':
                bang[hang][cot] = nguoi
                break
            else:
                print("O nay da duoc danh! Vui long chon o khac.\n")
        except ValueError:
            print("Vui long chi nhap so nguyen!\n")

# --- Chuong trinh chinh ---
if __name__ == "__main__":
    # Khoi tao bang trong
    bang_hien_tai = [
        [' ', ' ', ' '],
        [' ', ' ', ' '],
        [' ', ' ', ' ']
    ]

    print("=== CHAO MUNG DEN VOI TRO CHOI TIC-TAC-TOE ===")
    print("Ban la 'O' (Di truoc) - May la 'X' (Di sau)")
    in_bang(bang_hien_tai)

    while True:
        # 1. Luot cua nguoi choi
        print("\n--- LUOT CUA BAN ---")
        luot_nguoi_choi(bang_hien_tai)
        in_bang(bang_hien_tai)

        if kiem_tra_thang(bang_hien_tai, nguoi):
            print("=> CHUC MUNG! BAN DA THANG!")
            break
        if het_cho_trong(bang_hien_tai):
            print("=> TRO CHOI KET THUC VOI KET QUA HOA!")
            break

        # 2. Luot cua may
        print("\n--- LUOT CUA MAY ---")
        print("May dang suy nghi...")
        nuoc_di_may = tim_nuoc_di_tot_nhat(bang_hien_tai)
        bang_hien_tai[nuoc_di_may[0]][nuoc_di_may[1]] = may
        print(f"May da danh vao o: Hang {nuoc_di_may[0]}, Cot {nuoc_di_may[1]}")
        in_bang(bang_hien_tai)

        if kiem_tra_thang(bang_hien_tai, may):
            print("=> MAY DA THANG! CHUC BAN MAY MAN LAN SAU.")
            break
        if het_cho_trong(bang_hien_tai):
            print("=> TRO CHOI KET THUC VOI KET QUA HOA!")
            break
```

```python
may = 'X'
nguoi = 'O'

def in_bang(bang):
    print("-------------")
    for i in range(3):
        print(f"| {bang[i][0]} | {bang[i][1]} | {bang[i][2]} |")
        print("-------------")

def kiem_tra_thang(bang, nguoi_choi):
    # Kiem tra hang va cot
    for i in range(3):
        if bang[i][0] == bang[i][1] == bang[i][2] == nguoi_choi:
            return True
        if bang[0][i] == bang[1][i] == bang[2][i] == nguoi_choi:
            return True

    # Kiem tra duong cheo
    if bang[0][0] == bang[1][1] == bang[2][2] == nguoi_choi:
        return True
    if bang[0][2] == bang[1][1] == bang[2][0] == nguoi_choi:
        return True

    return False

def het_cho_trong(bang):
    for hang in bang:
        if ' ' in hang:
            return False
    return True

def tinh_diem(bang, do_sau):
    if kiem_tra_thang(bang, may):
        return 10 - do_sau
    if kiem_tra_thang(bang, nguoi):
        return do_sau - 10
    return 0

def alpha_beta(bang, do_sau, alpha, beta, la_luot_may):
    diem = tinh_diem(bang, do_sau)

    # Neu da co ket qua hoac het o trong
    if diem != 0 or het_cho_trong(bang):
        return diem

    if la_luot_may:
        gia_tri_max = -1000
        for i in range(3):
            for j in range(3):
                if bang[i][j] == ' ':
                    bang[i][j] = may
                    bien_so = alpha_beta(bang, do_sau + 1, alpha, beta, False)
                    bang[i][j] = ' '
                    gia_tri_max = max(gia_tri_max, bien_so)
                    alpha = max(alpha, gia_tri_max)
                    # Cat tia Alpha
                    if beta <= alpha:
                        break
            if beta <= alpha:
                break
        return gia_tri_max
    else:
        gia_tri_min = 1000
        for i in range(3):
            for j in range(3):
                if bang[i][j] == ' ':
                    bang[i][j] = nguoi
                    bien_so = alpha_beta(bang, do_sau + 1, alpha, beta, True)
                    bang[i][j] = ' '
                    gia_tri_min = min(gia_tri_min, bien_so)
                    beta = min(beta, gia_tri_min)
                    # Cat tia Beta
                    if beta <= alpha:
                        break
            if beta <= alpha:
                break
        return gia_tri_min

def tim_nuoc_di_tot_nhat(bang):
    gia_tri_tot_nhat = -1000
    nuoc_di = (-1, -1)

    # Alpha ban dau la cực am, Beta ban dau la cực duong
    alpha = -1000
    beta = 1000

    for i in range(3):
        for j in range(3):
            if bang[i][j] == ' ':
                bang[i][j] = may
                gia_tri_nuoc_di = alpha_beta(bang, 0, alpha, beta, False)
                bang[i][j] = ' '

                if gia_tri_nuoc_di > gia_tri_tot_nhat:
                    gia_tri_tot_nhat = gia_tri_nuoc_di
                    nuoc_di = (i, j)
    return nuoc_di

def choi_game():
    bang = [[' ' for _ in range(3)] for _ in range(3)]
    print("=== TIC TAC TOE: ALPHA-BETA PRUNING ===")
    print("Ban la 'O', May la 'X'. Ban di truoc!")
    in_bang(bang)

    while True:
        # Luot cua nguoi
        while True:
            try:
                h = int(input("Nhap hang (0-2): "))
                c = int(input("Nhap cot (0-2): "))
                if bang[h][c] == ' ':
                    bang[h][c] = nguoi
                    break
                else:
                    print("O da co nguoi danh!")
            except (ValueError, IndexError):
                print("Nhap sai! Vui long nhap lai.")

        in_bang(bang)
        if kiem_tra_thang(bang, nguoi):
            print("Ban thang!")
            break
        if het_cho_trong(bang):
            print("Hoa roi!")
            break

        # Luot cua may
        print("\nMay dang tinh toan...")
        n_di = tim_nuoc_di_tot_nhat(bang)
        bang[n_di[0]][n_di[1]] = may
        in_bang(bang)

        if kiem_tra_thang(bang, may):
            print("May thang!")
            break
        if het_cho_trong(bang):
            print("Hoa roi!")
            break


if __name__ == "__main__":
    choi_game()
```
