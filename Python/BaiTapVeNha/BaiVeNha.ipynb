```python
#bai1
def tron_hai_mang(mang_a, mang_b):
    mang_ket_qua = []
    do_dai_a = len(mang_a)
    do_dai_b = len(mang_b)

    # Trộn các phần tử chung, lấy giá trị nhỏ nhất
    do_dai_min = min(do_dai_a, do_dai_b)
    for chi_so in range(do_dai_min):
        mang_ket_qua.append(min(mang_a[chi_so], mang_b[chi_so]))

    # Thêm các phần tử còn lại của mảng a (nếu có)
    if do_dai_a > do_dai_min:
        mang_ket_qua.extend(mang_a[do_dai_min:])

    # Thêm các phần tử còn lại của mảng b (nếu có)
    elif do_dai_b > do_dai_min:
        mang_ket_qua.extend(mang_b[do_dai_min:])

    return mang_ket_qua
mang_a = [3, 9, 1, 4]
mang_b = [2, 7, 4, 3, 2, 8]

mang_moi = tron_hai_mang(mang_a, mang_b)
print(f"Mảng a: {mang_a}")
print(f"Mảng b: {mang_b}")
print(f"Mảng kết quả: {mang_moi}")
```

```python
#bai2
import math

# a. Tính tổng tam giác trên của đường chéo phụ (kể cả đường chéo phụ)
def tong_tam_giac_tren_phu(a, n):
    return sum(a[i][j] for i in range(n) for j in range(n - i))

# b. Chuyển phần tử âm thành trị tuyệt đối
def tri_tuyet_doi(a, m, n):
    for i in range(m):
        for j in range(n):
            a[i][j] = abs(a[i][j])

# c. Thay phần tử chẵn bằng x
def thay_chan_bang_x(a, m, n, x):
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 == 0:
                a[i][j] = x

# d. Kiểm tra mảng toàn chẵn
def la_toan_chan(a, m, n):
    return all(a[i][j] % 2 == 0 for i in range(m) for j in range(n))

# e. Kiểm tra ma trận đối xứng (a[i][j] == a[j][i])
def la_doi_xung(a, n):
    return all(a[i][j] == a[j][i] for i in range(n) for j in range(i))

# f. Kiểm tra đường chéo chính tăng dần
def duong_cheo_chinh_tang(a, n):
    return all(a[i][i] < a[i+1][i+1] for i in range(n - 1))

# g. Xuất các phần tử thuộc tam giác dưới đường chéo phụ (kể cả đường chéo phụ)
def xuat_tam_giac_duoi_phu(a, n):
    for i in range(n):
        dong = [str(a[i][j]) for j in range(n - 1 - i, n)]
        print(" ".join(dong))

# h. Kiểm tra đường chéo phụ giảm dần
def duong_cheo_phu_giam(a, n):
    return all(a[i][n-1-i] > a[i+1][n-2-i] for i in range(n - 1))

# VD
ma_tran = [
    [1, 2, 3],
    [2, 5, 6],
    [3, 6, 9]
]
n = 3
m = 3

print(f"a. Tong tam giac tren phu: {tong_tam_giac_tren_phu(ma_tran, n)}")
print(f"d. Toan chan: {la_toan_chan(ma_tran, m, n)}")
print(f"e. Doi xung: {la_doi_xung(ma_tran, n)}")
print(f"f. Duong cheo chinh tang: {duong_cheo_chinh_tang(ma_tran, n)}")
print(f"h. Duong cheo phu giam: {duong_cheo_phu_giam(ma_tran, n)}")
print("g. Cac phan tu tam giac duoi phu:")
xuat_tam_giac_duoi_phu(ma_tran, n)

# thay đổi giá trị
tri_tuyet_doi(ma_tran, m, n)
thay_chan_bang_x(ma_tran, m, n, 99)
print("Ma tran sau khi thay doi (b, c):", ma_tran)
```
