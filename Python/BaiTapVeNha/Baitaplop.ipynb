```python
#bai13
def tron_hai_mang(a, b):
    ket_qua = []
    n1 = len(a)
    n2 = len(b)
    min_length = min(n1, n2)
    for i in range(min_length):
        ket_qua.append(a[i] + b[i])
    if n1 > n2:
        ket_qua.extend(a[n2:])
    elif n2 > n1:
        ket_qua.extend(b[n1:])

    return ket_qua
mang_a = [3, 9, 1, 4]
mang_b = [2, 7, 4, 3, 2, 8]

mang_ket_qua = tron_hai_mang(mang_a, mang_b)
print("Mảng kết quả:", mang_ket_qua)
```

```python
#bai14
#a. Tạo mảng a chứa các số nguyên ngẫu nhiên.
import random

def create_random_matrix(m, n, min_val=0, max_val=100):

    matrix = []
    for i in range(m):
        dong = [random.randint(min_val, max_val) for i in range(n)]
        matrix.append(dong)
    return matrix
m_dong = int(input('Nhập số dòng (M): '))
n_cot = int(input('Nhập số cột (N): '))
random_matrix_a = create_random_matrix(m_dong, n_cot)
print(f"Ma trận a với {m_dong} dòng và {n_cot} cột:")
for dong in random_matrix_a:
    print(dong)
#b. Xuất các phần tử thuộc dòng k.
k_dong = int(input('Nhập số dòng muốn xuất  '))
if 1 <= k_dong <= m_dong:
    print(f"Các phần tử thuộc dòng {k_dong}:")
    print(random_matrix_a[k_dong-1])
else:
    print("Số dòng k không hợp lệ.")
#c. Xuất các phần tử thuộc cột k.
k_cot = int(input('Nhập số cột muốn xuất  '))
if 1 <= k_cot <= n_cot:
    print(f"Các phần tử thuộc cột {k_cot}:")
    column = [dong[k_cot-1] for dong in random_matrix_a]
    print(column)
else:
    print("Số cột k không hợp lệ.")
#d.Tìm dòng có tổng lớn nhất nhất 45.
sum_dong=0
k_dong_sum = int(input('Nhập số dòng muốn cộng  '))
if 1 <= k_dong_sum <= m_dong:
  sum_dong = sum(random_matrix_a[k_dong_sum-1])
  print(f"Tổng các phần tử thuộc dòng {k_dong_sum}: {sum_dong}")

#e. Tìm cột có tích nhỏ nhất.
min_nhonhat = float('inf')

for j in range(n_cot):
    khoi_tao = 1
    for i in range(m_dong):
        khoi_tao *= random_matrix_a[i][j]
    if khoi_tao < min_nhonhat:
        min_nhonhat = khoi_tao
        khoi_tao = j + 1

print(f"Cột có tích nhỏ nhất là: {min_nhonhat}.")

#f.Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a.
mang_f = []
for i in range(m_dong):

    if (i + 1) % 2 == 0:
        for j in range(n_cot):

            if (j + 1) % 2 != 0:
                mang_f.append(random_matrix_a[i][j])

print("Các phần tử thuộc dòng chẵn và cột lẻ:", mang_f)
#g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.
mang_g = []

for i in range(0, len(random_matrix_a), 2):
    for val in random_matrix_a[i]:
        if val % 2 == 0:
           mang_g.append(val)

if mang_g:
    tbc = sum(mang_g) / len(mang_g)
    print(f"\nCác phần tử chẵn ở dòng lẻ: {mang_g}")
    print(f"Trung bình cộng: {tbc}")
else:
    print("\nKhông tìm thấy phần tử chẵn nào ở các dòng lẻ.")
#h. Tính trung bình cộng các phần tử thuộc biên.
m = len(random_matrix_a)
n = len(random_matrix_a[0])
mang_bien = []

for i in range(m):
    for j in range(n):
        if i == 0 or i == m - 1 or j == 0 or j == n - 1:
            mang_bien.append(random_matrix_a[i][j])
if mang_bien:
    tbc_bien = sum(mang_bien) / len(mang_bien)
    print(f"Các phần tử thuộc biên: {mang_bien}")
    print(f"Trung bình cộng phần tử biên: {tbc_bien:.2f}")
else:
    print("Mảng rỗng.")
#i. Tính trung bình tích các phần tử không thuộc biên.
m = len(random_matrix_a)
n = len(random_matrix_a[0])

mang_loi = []
tich_loi = 1
for i in range(1, m - 1):
    for j in range(1, n - 1):
        val = random_matrix_a[i][j]
        mang_loi.append(val)
        tich_loi *= val
if mang_loi:
    trung_binh_tich = tich_loi / len(mang_loi)
    print(f"Các phần tử lõi (không thuộc biên): {mang_loi}")
    print(f"Tích các phần tử lõi: {tich_loi}")
    print(f"Trung bình tích: {trung_binh_tich:.2f}")
else:
    print("Ma trận quá nhỏ, không có phần tử nằm ngoài biên (lõi).")
```

```python
#bai15

import datetime

# a. Xây dựng cấu trúc SINHVIEN
class SinhVien:
    def __init__(self, ma_sv, ho_ten, nam_sinh, diem_tb):
        self.ma_sv = str(ma_sv)[:10]      # Giới hạn 10 ký tự
        self.ho_ten = str(ho_ten)[:20]    # Giới hạn 20 ký tự
        self.nam_sinh = int(nam_sinh)
        self.diem_tb = float(diem_tb)

# b. Tạo danh sách n sinh viên (Dữ liệu mẫu)
danh_sach_sv = [
    SinhVien("01DH0001", "Phan Thanh Lan", 2004, 8.5),
    SinhVien("02DH0002", "Nguyen Van An", 2007, 4.5),
    SinhVien("03CD0003", "Phan Hoang Nam", 2006, 6.0),
    SinhVien("04DH0004", "Tran Thi Lan", 2005, 3.0),
    SinhVien("05KT0005", "Le Huu Loc", 2002, 7.2)
]

# c. Đếm sinh viên đủ điều kiện lên lớp (DTB >= 5)
def dem_sv_len_lop(ds):
    # Dùng list comprehension để đếm cho nhanh
    ket_qua = [sv for sv in ds if sv.diem_tb >= 5]
    return len(ket_qua)

# d. Xuất các sinh viên đủ 20 tuổi
def xuat_sv_20_tuoi(ds):
    nam_hien_tai = datetime.datetime.now().year # Năm 2026 theo hệ thống
    print(f"\n--- Sinh viên từ 20 tuổi trở lên ({nam_hien_tai}): ---")
    for sv in ds:
        tuoi = nam_hien_tai - sv.nam_sinh
        if tuoi >= 20:
            print(f"Mã: {sv.ma_sv} | Tên: {sv.ho_ten} | Tuổi: {tuoi}")

# e. Đếm sinh viên hệ Đại học (Mã chứa "DH" ở vị trí 2, 3)
# Lưu ý: Vị trí 2,3 trong chuỗi tương ứng index 2 và 3 trong Python
def dem_sv_he_dh(ds):
    dem = 0
    for sv in ds:
        # Cắt chuỗi từ index 2 đến 4 (lấy 2 và 3)
        if sv.ma_sv[2:4] == "DH":
            dem += 1
    return dem

# f. Đếm sinh viên có tên "Lan"
def dem_ten_lan(ds):
    dem = 0
    for sv in ds:
        # split() tách chuỗi theo khoảng trắng, [-1] lấy từ cuối cùng
        ten_rieng = sv.ho_ten.split()[-1]
        if ten_rieng.lower() == "lan":
            dem += 1
    return dem

# g. Đếm sinh viên có họ "Phan"
def dem_ho_phan(ds):
    dem = 0
    for sv in ds:
        # startswith kiểm tra chuỗi bắt đầu bằng...
        if sv.ho_ten.startswith("Phan"):
            dem += 1
    return dem

# --- CHẠY KIỂM TRA ---
print(f"1. Số SV đủ điều kiện lên lớp: {dem_sv_len_lop(danh_sach_sv)}")

xuat_sv_20_tuoi(danh_sach_sv)

print(f"\n2. Số SV hệ Đại học (DH): {dem_sv_he_dh(danh_sach_sv)}")
print(f"3. Số SV tên 'Lan': {dem_ten_lan(danh_sach_sv)}")
print(f"4. Số SV họ 'Phan': {dem_ho_phan(danh_sach_sv)}")
```
